```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

df = pd.read_csv('county_level_df')
cols=[c for c in [c for c in df.columns if not c.startswith('geo')] if not c.startswith('net')]
#df[cols].to_csv('df_for_josh')


var = 'corona_deaths'#'fitted_deaths'#'corona_deaths'

df1=df#[df['corona_deaths'] > 0]
print(df1.shape)

f, ax = plt.subplots()

ax.loglog(df1[var],df1['corona_cases'],'o',alpha=0.5)

real_cases = df1[var].sort_values()

ax.fill_between(df1[var].sort_values(),real_cases/0.02,real_cases/0.006,color='r',alpha=0.3,
        label='expected cases\nfatality rate: 0.6% to 2%')

ax.loglog(df1[var].sort_values(),real_cases/0.006,'r--')
ax.loglog(df1[var].sort_values(),real_cases/0.02,'r--')

ax.set_xlabel('# deaths',fontsize=15)
ax.set_ylabel('# confirmed cases',fontsize=15)



plt.legend(fontsize=15,loc=2)

#plt.show()
plt.savefig('cases_vs_deaths_expected_deaths.pdf')
#plt.close()
plt.show()
#print(df1.shape)
```
